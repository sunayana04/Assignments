```python
class Dog:
    def __init__(self, name, age, coat_color):
        self.name = name
        self.age = age
        self.coat_color = coat_color

    def description(self):
        print("Name: ", self.name)
        print("Age: ", self.age)

    def get_info(self):
        print("Coat Color: ", self.coat_color)


class JackRussellTerrier(Dog):
    def __init__(self, name, age, coat_color, energy_level):
        super().__init__(name, age, coat_color)
        self.energy_level = energy_level

    def play_fetch(self):
        print("Playing fetch with high energy level.")

    def bark(self):
        print("Barking loudly with excitement.")


class Bulldog(Dog):
    def __init__(self, name, age, coat_color, size):
        super().__init__(name, age, coat_color)
        self.size = size

    def snore(self):
        print("Snoring loudly due to heavy build.")

    def drool(self):
        print("Drooling due to short snout and loose skin.")


# Create objects of Dog class
dog1 = Dog("Buddy", 3, "Brown")
dog2 = Dog("Charlie", 5, "White")

# Call description() and get_info() methods of Dog class
dog1.description()
dog1.get_info()

dog2.description()
dog2.get_info()

# Create objects of JackRussellTerrier class
jack_russell1 = JackRussellTerrier("Rocky", 2, "Tan and White", "High")
jack_russell2 = JackRussellTerrier("Max", 4, "Black and White", "Medium")

# Call description(), get_info(), play_fetch(), and bark() methods of JackRussellTerrier class
jack_russell1.description()
jack_russell1.get_info()
jack_russell1.play_fetch()
jack_russell1.bark()

jack_russell2.description()
jack_russell2.get_info()
jack_russell2.play_fetch()
jack_russell2.bark()

# Create objects of Bulldog class
bulldog1 = Bulldog("Rocky", 3, "Brown", "Medium")
bulldog2 = Bulldog("Duke", 6, "Fawn", "Large")

# Call description(), get_info(), snore(), and drool() methods of Bulldog class
bulldog1.description()
bulldog1.get_info()
bulldog1.snore()
bulldog1.drool()

bulldog2.description()
bulldog2.get_info()
bulldog2.snore()
bulldog2.drool()
```
